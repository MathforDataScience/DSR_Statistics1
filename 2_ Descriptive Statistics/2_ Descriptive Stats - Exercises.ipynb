**Exercise 3:**

Question:
Compute the variance of the numbers.
a) Use population variance
b) Use sample variance

ages = [21, 23, 25, 27, 29, 31, 33]

```python
import numpy as np
ages = np.array([21, 23, 25, 27, 29, 31, 33])
pop_variance = np.var(ages)
pop_variance
```

b)
Let's say we take a sample of 5 ages from the original group:
sample_ages = [21, 23, 27, 31, 33]

```python
sample_ages = np.array([21, 23, 27, 31, 33])
sample_variance = np.var(sample_ages, ddof=1)

print('Sample Variance:', sample_variance)
```

**Exercise 6a:**

show with python plots for bar chart, histogram, line diagram, 
pie diagram in a 2x2 subfigure matrix

```python
import matplotlib.pyplot as plt
import numpy as np

# Some data
x = np.arange(0., 10., 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 15, 5, 20, 10]
hist_data = np.random.randn(100)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Bar chart
axs[0, 0].bar(categories, values)
axs[0, 0].set_title('Bar chart')

# Histogram
axs[0, 1].hist(hist_data, bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Histogram')

# Line chart
axs[1, 0].plot(x, y1, label='sin(x)')
axs[1, 0].plot(x, y2, label='cos(x)', linestyle='--')
axs[1, 0].set_title('Line diagram')
axs[1, 0].legend()

# Pie chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[1, 1].set_title('Pie chart')

# Adjust the spacing between the plots
plt.tight_layout()
plt.show()
```

**Exercise 6b:**

show with python plots in a 2x2 subfigure matrix for Scatter plot, Box plot (Box-and-whisker plot), Heatmap, bubble chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some data
np.random.seed(0)  # For reproducibility

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.randn(100)

sizes = np.random.rand(50) * 1000  # For bubble chart
matrix = np.random.rand(10, 12)  # For heatmap

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Scatter plot
axs[0, 0].scatter(x, y)
axs[0, 0].set_title('Scatter plot')
axs[0, 0].grid()

# Box plot
axs[0, 1].boxplot(z)
axs[0, 1].set_title('Box plot')

# Heatmap
sns.heatmap(matrix, ax=axs[1, 0], cmap='viridis')
axs[1, 0].set_title('Heatmap')

# Bubble chart
bubble = axs[1, 1].scatter(x, y, s=sizes, alpha=0.5)
axs[1, 1].set_title('Bubble chart')
axs[1, 1].grid()

# Layout adjustment
fig.tight_layout()

plt.show()
```
