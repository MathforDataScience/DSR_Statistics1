```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f
```

```python
# Define degrees of freedom for different scenarios
dfn_values = [2, 3, 5]  # degrees of freedom numerator (between groups)
dfd_values = [18, 45, 100]  # degrees of freedom denominator (within groups)
```

```python
# Generate data for x axis
x = np.linspace(0, 5, 1000)

# Create the plot
fig, ax = plt.subplots()

# Plot the density function for each scenario
for dfn, dfd in zip(dfn_values, dfd_values):
    # Generate data for y axis
    y = f.pdf(x, dfn, dfd)
    
    # Plot the density function
    ax.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')

# Set the title and labels
ax.set_title('Density Functions for F Distributions with Different Degrees of Freedom')
ax.set_xlabel('Value')
ax.set_ylabel('Density')

# Add legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```
