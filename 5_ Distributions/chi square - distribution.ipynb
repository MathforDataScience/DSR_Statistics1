```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom to plot
dfs = [1, 2, 3, 4, 5]

# Define the range of x (endpoints based on the highest df)
x = np.linspace(chi2.ppf(0.01, dfs[-1]), chi2.ppf(0.99, dfs[-1]), 100)

# Create the plot
plt.figure(figsize=(10, 6))

for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f'df={df}')

plt.title('Chi-square Distribution PDF for Different Degrees of Freedom')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

plt.grid()
plt.show()
```
